```python
#test code out of the framework (personal practice)
from datetime import datetime

def check_class_status(is_holiday):

    today = datetime.now().strftime("%A")

    if today == 'Monday':
        if is_holiday:
            return "Yes, it's a holiday. There is no class today."
        else:
            return "No, it's not a holiday. Yes, there is a class today."
    else:
        return "Today is not Monday."


is_holiday = True
print(check_class_status(is_holiday))
```
